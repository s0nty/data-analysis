```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
co=pd.read_csv("latest Covid-19 India Status1.csv")
co
```

# Adding a new coloums to find the number of patients who has been recoverd from covid-19

```python
co['Recovered']=co['Active']+co['Deaths']
```

```python
 co['Recovered']=co['Total Cases']-co['Recovered']
```

```python
co['Recoverd Ratio']=co['Recovered']/co['Total Cases']*100
```

```python
co.columns=['State', 'Total Cases', 'Active', 'Deaths', 'Active_Ratio',
       'Death Ratio','Recovered', 'Recoverd Ratio']
```

# with the help of barplot we can see in which state has the highest ratio of active cases

```python

import seaborn as sns

sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(co['State'],co['Active_Ratio'],data=co)
plt.xticks(rotation=90)
plt.show()
```

# in this we find top 5 states who has the highest recovery rate 

```python
 c=co.nlargest(5,'Recoverd Ratio')
c
```

```python
plt.figure(figsize=(8,8))
# labels = co.index
plt.pie(x='Recoverd Ratio',data=c ,labels='State',autopct='%.2f%%')
plt.title("Recoverd Ratio in Top 5 states in India", fontsize = 24) 
plt.show()

```

# last 5 states with bad recovery rate

```python
b=co.nsmallest(5,'Recoverd Ratio')
b
```

```python
plt.pie(x='Recoverd Ratio',data=b ,labels='State',autopct='%.2f%%')
plt.title("Recoverd Ratio in Last 5 states in India", fontsize = 24) 
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(x='State',y='Recoverd Ratio',data=co)
plt.xticks(rotation=90)
plt.show()
```
